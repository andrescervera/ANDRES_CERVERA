```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Crear clusters ficticios con 2 características y 3 clusters
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.5, random_state=42)

# 1. Importar el modelo
model = KMeans(n_clusters=3)

# 2. Entrenar el modelo
model.fit(X)

# 3. Predecir
predictions = model.predict(X)

# Plotear los puntos con los clusters coloreados
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=predictions, cmap='viridis', alpha=0.5)
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], c='red', s=100, marker='X', label='Centroides')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.title('Clusters KMeans')
plt.legend()
plt.grid(True)
plt.show()
```

```python
inertia = model.inertia_
print("Inercia del modelo KMeans:", inertia)
```

En este caso, podemos sacar el accuracy. ¿Podremos en un dataset real de clustering?

```python
centers = model.cluster_centers_
print("Centros de los clusters:", centers)
```

# ¿Algo interesante respecto a los centros?

```python
import matplotlib.pyplot as plt

# Crear el scatter plot con los puntos coloreados por cluster
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=predictions, cmap='viridis', alpha=0.5)
plt.scatter(centers[:, 0], centers[:, 1], c='yellow', marker='X', s=100, label='Centros de clusters')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.title('Clusters KMeans con Centros')
plt.colorbar(label='Cluster')
plt.legend()
plt.grid(True)
plt.show()
```
